```
import cv2
cv2.imread(file_name,flag) -> 이미지를 읽어 numpy 객체로 만드는 함수
```

- file_name : 읽고자 하는 파일

- flag : 이미지를 읽는 방법 설정

imread_color : 이미지를 color로 읽고, 투명한 부분은 무시

imread_grayscale : 이미지를 grayscale(흑백)으로 읽기

imread_unchanged : 이미지를 color로 읽고, 투명한 부분도 읽기 (Alpha)

-> 반환 값 : numpy 객체 (행,열, 색상:기본 BGR)
일반적으로 RGB이나 OpenCV에서는 RGB로 순서가 다른 점 참고

```
cv2.imshow(title,image) # 특정한 이미지를 화면에 출력
cv2.mwrite(file_name,image) # 특정한 이미지를 파일로 저장하는 함수
```

cv2.cvtColor로 로드했을 때, 앞서 언급했듯 RGB가 아닌 BGR로 변환되기 때문에 plt로 출력할 때 BGR2RGB(BGR to RGB)로 변환해서 출력해야 한다.

마찬가지로 BGR2GRAY라면 GRAY2RGB로 하는 습관을 들이자.

```
import cv2
import matplotlib.pyplot as plt # Colab은 윈도우 창을 띄울 수 없기 때문에 matplot의 imshow를 써야함

img_basic=cv2.imread('images.jfif',cv2.IMREAD_COLOR) # 대문자로 해줘야 작동
plt.imshow(cv2.cvtColor(img_basic,cv2.COLOR_BGR2RGB))
plt.show()

img_basic=cv2.imread('images.jfif',cv2.IMREAD_GRAYSCALE)
plt.imshow(cv2.cvtColor(img_basic,cv2.COLOR_GRAY2RGB))
plt.show()

img_basic=cv2.imread('images.jfif',cv2.IMREAD_GRAYSCALE) 
plt.imshow(cv2.cvtColor(img_basic,cv2.COLOR_GRAY2RGB))
plt.show()

img_basic=cv2.imread('images.jfif',cv2.IMREAD_UNCHANGED) 
plt.imshow(cv2.cvtColor(img_basic,cv2.COLOR_BGR2RGB))
plt.show()
```
