```
import cv2
image=cv2.imread('images.jfif')

# 픽셀 수 및 이미지 크기 확인
print(image.shape) # (174,289,3) 이 때 3은 각각의 BGR을 나타내는듯 
print(image.size)

# 이미지 numpy 객체의 특정 픽셀을 가리키기
px=image[100,100]

# B,G,R 순서로 출력된다.
# 단, gray scale인 경우 B,G,R로 구분되지 않는다
print(px)

# R 값만 출력
print(px[2])
```

```
import cv2
import matplotlib.pyplot as plt
import time

image=cv2.imread('images.jfif')

start_time=time.time()
for i in range(100):
  for j in range(100):
    image[i,j]=[255,255,255] # 흰색
print("--- %s seconds ---" % (time.time() - start_time))

plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.show()

start_time=time.time()
image[0:100,0:100]=[0,0,0] # 검은색
print("--- %s seconds ---" % (time.time() - start_time))
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.show()
```

- 이렇게 image 내의 numpy 배열의 요소를 직접 건드려서 이미지 처리를 할 수 있다.

- 하지만 시간초를 통해 알 수 있듯이 for문으로 처리하는 것보다 한 번에 리스트로 처리하는게 더 빠르다

ROI 추출 및 복사
- Region of interest : 이미지 상에서 내가 관심있어하는 일부 영역

```
import cv2
import matplotlib.pyplot as plt

image=cv2.imread('images.jfif')

roi=image[30:80,150:200]

image[0:50,0:50]=roi
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.show() # 귀엽다 ㅋㅋㅋㅋ
```
