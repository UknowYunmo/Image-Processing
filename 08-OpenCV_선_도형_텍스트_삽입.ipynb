근데 좀 어이없는건 image에 선을 그은 걸 각각 image1, image2, image3에 저장했는데 이러한 변화들이 image에도 똑같이 나타나서 사실상 선언을 따로 하는게 의미가 없는 것이 확인되었다.

그렸던 선을 지우는 함수는 별도로 없기 때문에(사실상 선 개념이 아니고 선처럼 보일뿐이니), 매번 새롭게 돌아가서 추가하고 싶다면 .copy()를 계속 써서 그 위에 하거나 아니면 새로 계속 만들어거나 해야하는 것 같다.

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

image=np.full((512,512,3),255,np.uint8) # 8bit 최대 255까지의 값을 가질 수 있는 512x512 픽셀 + 3(RGB) 이미지 생성
plt.imshow(image)
plt.show()

image1=cv2.line(image,(0,0),(255,255),(255,0,0),7)
plt.imshow(image1)
plt.show()

image=np.full((512,512,3),255,np.uint8)
image2=cv2.line(image,(0,0),(255,255),(0,255,0),2)
plt.imshow(image2)
plt.show()

image=np.full((512,512,3),255,np.uint8)
image3=cv2.line(image,(0,0),(255,255),(0,0,255),10)
plt.imshow(image3)
plt.show()

image=np.full((512,512,3),255,np.uint8)
plt.imshow(image)
plt.show()
```

```
cv2.rectangle(image,start,end,color,thickness) # 하나의 사각형을 그리는 함수
cv2.circle(image,center,radian,color,thickness) # 하나의 원을 그리는 함수
cv2.polylines(image,points,is_closed,color,thickness) # 하나의 다각형을 그리는 함수
```

- start : 시작 좌표 (2차원)
- end : 종료 좌표 (2차원)
- thickness : 선의 두께 (채우기: -1)
---
- center : 원의 중심 (2차원)
- radian : 반지름
- thickness : 선의 두께 (채우기: -1)
---
- points : 꼭지점들
- is_closed : 닫힌 도형 여부
- thickness : 선의 두께 (채우기: -1)

```
image=np.full((512,512,3),255,np.uint8)
cv2.rectangle(image,(100,100),(300,300),(256,0,0),3)
plt.imshow(image)
plt.show()
```

```
image=np.full((512,512,3),255,np.uint8)
cv2.circle(image,(200,200),100,(120,150,150),-1)
plt.imshow(image)
plt.show()
```

```
image=np.full((512,512,3),255,np.uint8)
points=np.array([[30,30],[128,258],[483,444],[400,150]])
cv2.polylines(image,[points],True,(0,0,255),4)
plt.imshow(image)
plt.show()
```

```
cv2.putText(image,text,position,font_typefont_scale,color) # 하나의 텍스트를 그리는 함수
```

- position : 텍스트가 출력될 위치
- font_type : 글씨체
- font_scale : 글씨 크기 가중치

```
image=np.full((512,512,3),255,np.uint8)
cv2.putText(image,'OpenCV Fighting!',(100,200),cv2.FONT_ITALIC,0.8,(0,0,0))
plt.imshow(image)
plt.show()
```
